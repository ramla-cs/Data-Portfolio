### Visualization 1: Distribution of Volcano Elevations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Visualization 1: Distribution of Volcano Elevations - Histogram and Box Plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Histogram of volcano elevations
sns.histplot(df['Elevation'], bins=30, kde=True, ax=ax[0])
ax[0].set_title('Distribution of Volcano Elevations')
ax[0].set_xlabel('Elevation (m)')
ax[0].set_ylabel('Frequency')

# Box Plot of volcano elevations
sns.boxplot(y=df['Elevation'], ax=ax[1])
ax[1].set_title('Box Plot of Volcano Elevations')
ax[1].set_ylabel('Elevation (m)')

plt.tight_layout()
plt.show()
```

The histogram and kernel density estimate (KDE) show that most volcanoes have elevations ranging from low to moderate heights, with a significant tail extending towards higher elevations.
The box plot reveals a wide range of elevations among the volcanoes, with several outliers indicating volcanoes at exceptionally high elevations.

### Visualisation 3: Geospatial Analysis

   - Given that latitude and longitude data are available, I can carry out a geospatial analysis to visualize the distribution of volcanoes on a map.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import get_provider, Vendors
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import transform
import numpy as np
import pandas as pd

# Ensure output is displayed in the notebook
output_notebook()

# Convert lat/lon to Web Mercator coordinates
def wgs84_to_web_mercator(df, lon="Longitude", lat="Latitude"):
    k = 6378137
    df["Longitude"] = df[lon] * (k * np.pi/180.0)
    df["Latitude"] = np.log(np.tan((90 + df[lat]) * np.pi/360.0)) * k
    return df

# Assuming df is your DataFrame with 'Latitude', 'Longitude', and 'Elevation'
df = wgs84_to_web_mercator(df)

# Define the color mapper: linear mapping from elevations to a color palette
color_mapper = LinearColorMapper(palette="Viridis256", low=df['Elevation'].min(), high=df['Elevation'].max())

# Add elevation color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title='Elevation')

# Set up the figure
p = figure(title="Volcano Locations by Elevation", x_axis_type="mercator", y_axis_type="mercator",
           x_axis_label="Longitude", y_axis_label="Latitude", sizing_mode="scale_width", height=400)

# Add the map tile using get_provider
tile_provider = get_provider(Vendors.CARTODBPOSITRON)
p.add_tile(tile_provider)

# Add points using 'circle' glyphs, color by elevation
p.circle(x="Longitude", y="Latitude", source=ColumnDataSource(df),
         size=5, color=transform('Elevation', color_mapper), alpha=0.6)

# Add the color bar to the figure
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

## Visualisation 4: Exploring Temporal Patterns

   - Explore temporal patterns such as the frequency of eruptions over time.

```python
# Here I've rounded down to the nearest century for grouping, but could adjust for decades or specific years.
## df['Century'] = (df['Last_Eruption_Year'] // 100) * 100

# Adjust the dataset for analysis by grouping eruption years into more detailed categories
# Define a function to categorize eruption years
def categorize_eruption_year(year):
    if year < 0:
        return "Pre-AD"
    elif year < 1000:
        return "0-1000 AD"
    else:
        # Group by decade for years 1000 AD and later
        decade = (year // 10) * 10
        return f"{decade}s"

# Apply the categorization function to the 'Last_Eruption_Year' column
df['Eruption_Period'] = df['Last_Eruption_Year'].dropna().apply(categorize_eruption_year)

# Count the number of eruptions per categorized period
eruptions_per_period = df['Eruption_Period'].value_counts().sort_index()

# Plot
plt.figure(figsize=(14, 8))
eruptions_per_period.plot(kind='bar')
plt.title('Frequency of Eruptions Over Time')
plt.xlabel('Time Period')
plt.ylabel('Number of Eruptions')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# This adjustment will allow us to see a more nuanced view of eruption frequencies over time.


```

# Correlation Analysis

   - Explore correlations between variables such as latitude, longitude, and elevation to understand any relationships between these factors.

### Correlation Matrix

```python
# Calculate correlation matrix for latitude, longitude, and elevation
correlation_matrix = df[['Latitude', 'Longitude', 'Elevation']].corr()

# Display the correlation matrix
print(correlation_matrix)
```

### Heat Map
To better visualize these correlations, here is a heatmap, which provides a color-coded representation of the correlation matrix:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Latitude, Longitude, and Elevation')
plt.show()
```

This heatmap visually shows you the strength and direction of the relationships between latitude, longitude, and elevation.  I am specifically interested in seeing if there is a correlation between elevation and the other two variables, latitude and longitude.  There does not appear to be a strong correlation.
